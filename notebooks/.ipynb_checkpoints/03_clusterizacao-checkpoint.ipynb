Parte 3 - Clusterização

Q1 A - Realizar o agrupamento dos países em 3 grupos distintos. Para tal, use: K-Médias

```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_theme(palette="viridis")
pd.set_option("display.max_colwidth", 160)

data_path = Path("../data/country_data_scaled.csv")
df_scaled_full = pd.read_csv(data_path)
df_scaled = df_scaled_full.drop(columns=["country"])

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(df_scaled)

df_clusters = df_scaled_full.copy()
df_clusters["cluster_kmeans"] = kmeans.labels_

centroids = pd.DataFrame(kmeans.cluster_centers_, columns=df_scaled.columns)

pca = PCA(n_components=2)
pca_data = pca.fit_transform(df_scaled)
df_pca = pd.DataFrame(pca_data, columns=["PC1", "PC2"])
df_pca["cluster"] = kmeans.labels_

plt.figure(figsize=(10,6))
sns.scatterplot(data=df_pca, x="PC1", y="PC2", hue="cluster", palette="viridis", s=80)
plt.title("Visualização dos clusters - K-Médias (redução PCA)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(9,3.6))
sns.heatmap(centroids.round(2), annot=True, fmt=".2f", cmap="viridis", cbar=False,
            yticklabels=[0,1,2])
plt.title("Centróides (escala padronizada)")
plt.tight_layout()
plt.show()

X = df_scaled.to_numpy()
labels = kmeans.labels_
centers = kmeans.cluster_centers_
representantes = []
for c in np.unique(labels):
    idx = np.where(labels == c)[0]
    dist = np.linalg.norm(X[idx] - centers[c], axis=1)
    rep_idx = idx[np.argmin(dist)]
    representantes.append(df_scaled_full.iloc[rep_idx]["country"])

cols = ['child_mort','life_expec','total_fer','inflation','income','gdpp','exports','imports','health']
summary = df_clusters.groupby("cluster_kmeans")[cols].mean()

summary_meta = pd.DataFrame({
    "n_paises": df_clusters["cluster_kmeans"].value_counts().sort_index(),
    "pais_representativo": representantes
})
cluster_names = {0: "Intermediário", 1: "Alta Renda", 2: "Baixa Renda"}
summary_meta["nome"] = summary_meta.index.map(cluster_names)

summary_final = (
    summary_meta
    .join(summary)
    .round(2)
    .rename_axis("cluster")
    .reset_index()
    [["cluster","nome","n_paises","pais_representativo"] + cols]
)
display(summary_final)

samples = (
    df_clusters
    .groupby("cluster_kmeans")["country"]
    .apply(lambda s: ", ".join(s.head(8)))
    .rename("exemplos_de_paises")
    .reset_index()
    .rename(columns={"cluster_kmeans":"cluster"})
)
display(samples)
```

Q1 B - Realizar o agrupamento dos países em 3 grupos distintos. Para tal, use: Clusterização Hierárquica

```python
import pandas as pd
import numpy as np
from pathlib import Path
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_theme(palette="viridis")

data_path = Path("../data/country_data_scaled.csv")
df_scaled_full = pd.read_csv(data_path)
df_scaled = df_scaled_full.drop(columns=["country"])

linkage_matrix = linkage(df_scaled, method='ward')

plt.figure(figsize=(14, 6))
dendrogram(linkage_matrix, labels=df_scaled_full["country"].values, leaf_rotation=90, leaf_font_size=7)
plt.title("Dendrograma - Clusterização Hierárquica (Ward)")
plt.tight_layout()
plt.show()

clusters_hier = fcluster(linkage_matrix, 3, criterion='maxclust')

df_hier = df_scaled_full.copy()
df_hier["cluster_hier"] = clusters_hier

display(df_hier["cluster_hier"].value_counts().sort_index())

samples_hier = (
    df_hier.groupby("cluster_hier")["country"]
    .apply(lambda s: ", ".join(s.head(8)))
    .rename("exemplos_de_paises")
    .reset_index()
)
display(samples_hier)
```
