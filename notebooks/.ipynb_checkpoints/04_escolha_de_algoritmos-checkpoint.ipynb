Parte 4 - Escolha de algoritmos

Q2 - Refaça o algoritmo apresentado na questão 1 a fim de garantir que o cluster seja representado pelo dado mais próximo ao seu baricentro em todas as iterações do algoritmo:

```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.decomposition import PCA
from sklearn.metrics import pairwise_distances_argmin_min
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_theme(palette="viridis")
pd.set_option("display.max_colwidth", 160)

data_path = Path("../data/country_data_scaled.csv")
df_scaled_full = pd.read_csv(data_path)
df_scaled = df_scaled_full.drop(columns=["country"]).to_numpy()
cols = ['child_mort','life_expec','total_fer','inflation','income','gdpp','exports','imports','health']

np.random.seed(42)
k = 3
max_iter = 100
idx_init = np.random.choice(len(df_scaled), k, replace=False)
medoids = df_scaled[idx_init]

for _ in range(max_iter):
    dists = np.linalg.norm(df_scaled[:, None] - medoids, axis=2)
    labels_med = np.argmin(dists, axis=1)
    new_medoids = []
    for c in range(k):
        pts = df_scaled[labels_med == c]
        if len(pts) == 0:
            new_medoids.append(medoids[c])
            continue
        centroid = pts.mean(axis=0)
        nearest_idx, _ = pairwise_distances_argmin_min(centroid.reshape(1,-1), pts)
        new_medoids.append(pts[nearest_idx[0]])
    new_medoids = np.array(new_medoids)
    if np.allclose(new_medoids, medoids):
        break
    medoids = new_medoids

df_med = df_scaled_full.copy()
df_med["cluster_medoid"] = labels_med

pca = PCA(n_components=2)
pca_data = pca.fit_transform(df_scaled)
df_pca_med = pd.DataFrame(pca_data, columns=["PC1", "PC2"])
df_pca_med["cluster"] = labels_med

plt.figure(figsize=(10,6))
sns.scatterplot(data=df_pca_med, x="PC1", y="PC2", hue="cluster", palette="viridis", s=80)
plt.title("Visualização dos clusters - Medóides (redução PCA)")
plt.tight_layout()
plt.show()

summary_means = (
    pd.DataFrame(df_scaled, columns=df_scaled_full.drop(columns=["country"]).columns)
    .assign(cluster=df_med["cluster_medoid"].values)
    .groupby("cluster")[cols].mean()
)

plt.figure(figsize=(9,3.6))
sns.heatmap(summary_means.round(2), annot=True, fmt=".2f", cmap="viridis", cbar=False,
            yticklabels=summary_means.index.tolist())
plt.title("Perfil médio por cluster (Medóides, escala padronizada)")
plt.tight_layout()
plt.show()

rep_names = []
for c in range(k):
    m = medoids[c].reshape(1,-1)
    idx_all, _ = pairwise_distances_argmin_min(m, df_scaled)
    rep_names.append(df_scaled_full.iloc[idx_all[0]]["country"])

summary_meta = pd.DataFrame({
    "n_paises": df_med["cluster_medoid"].value_counts().sort_index(),
    "pais_representativo": rep_names
}).rename_axis("cluster").reset_index()

cluster_names = {0: "Intermediário (Medóide)", 1: "Alta Renda (Medóide)", 2: "Baixa Renda (Medóide)"}
summary_meta["nome"] = summary_meta["cluster"].map(cluster_names)

summary_final_med = (
    summary_meta
    .join(summary_means.round(2), on="cluster")
    [["cluster","nome","n_paises","pais_representativo"] + cols]
)
display(summary_final_med)

samples_med = (
    df_med.groupby("cluster_medoid")["country"]
    .apply(lambda s: ", ".join(s.head(8)))
    .rename("exemplos_de_paises")
    .reset_index()
    .rename(columns={"cluster_medoid":"cluster"})
)
display(samples_med)
```
