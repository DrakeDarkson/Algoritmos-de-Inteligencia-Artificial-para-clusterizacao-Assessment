Parte 2 - Escolha de base de dados

Q2 - Quantos países existem no dataset?

```python
import pandas as pd

df = pd.read_csv("../data/Country-data.csv")
df.columns = [c.strip().lower() for c in df.columns]

print("Dimensões do dataset:", df.shape)
print("Quantidade de países distintos:", df["country"].nunique())
display(df)
```

Q3 - Mostre através de gráficos a faixa dinâmica das variáveis que serão usadas nas tarefas de clusterização:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.style.use("ggplot")
sns.set_theme(palette="viridis")

num_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(14, 6))
sns.boxplot(data=num_df, orient="h")
plt.title("Faixa dinâmica das variáveis (boxplot)")
plt.tight_layout()
plt.show()

num_df.hist(bins=20, figsize=(14, 10), edgecolor="black")
plt.suptitle("Distribuição das variáveis numéricas (histogramas)", y=1.02)
plt.tight_layout()
plt.show()
```

Q4 - Realize o pré-processamento adequado dos dados:

```python
from sklearn.preprocessing import StandardScaler

X = num_df.fillna(num_df.median(numeric_only=True))

sk = X.skew(numeric_only=True)
cols_log = [c for c in X.columns if sk[c] > 1.0 and X[c].min() > 0]
X[cols_log] = np.log1p(X[cols_log])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
df_scaled = pd.DataFrame(X_scaled, columns=X.columns)
df_scaled_full = pd.concat([df["country"], df_scaled], axis=1)

print("Colunas transformadas com log1p:", cols_log)
print("Shape após pré-processamento:", df_scaled_full.shape)
display(df_scaled_full.head())

df_scaled_full.to_csv("../data/country_data_scaled.csv", index=False)
print("Arquivo pré-processado salvo em: ../data/country_data_scaled.csv")
```
