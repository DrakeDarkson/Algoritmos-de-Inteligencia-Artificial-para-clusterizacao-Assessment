```python
import sys, os, platform, subprocess
print("Python:", sys.version)
print("Executável:", sys.executable)
print("Plataforma:", platform.platform())
print("VIRTUAL_ENV:", os.environ.get("VIRTUAL_ENV"))
print("CONDA_PREFIX:", os.environ.get("CONDA_PREFIX"))

print("\nPacotes principais:")
import numpy, pandas, scipy, sklearn, matplotlib
print("numpy", numpy.__version__)
print("pandas", pandas.__version__)
print("scipy", scipy.__version__)
print("scikit-learn", sklearn.__version__)
print("matplotlib", matplotlib.__version__)

print("\nTrecho do pip freeze:")
freeze = subprocess.check_output(["pip","freeze"], text=True)
print("\n".join(freeze.splitlines()[:30]))
```
